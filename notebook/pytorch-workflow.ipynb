# **The 5-Step Training Loop**

Every deep learning model follows this cycle:

1. forward
2. compute loss
3. backward
4. optimizer step
5. zero gradients

```python
import torch
torch.__version__
```

```python
x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y = torch.tensor([[3.0], [5.0], [7.0], [9.0]])

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

learning_rate = 0.01
```

## Training from scratch.
purely PyTorch

```python
for epoch in range(100):
    # 1. Forward
    y_pred = w * x + b

    # 2. Loss
    loss = ((y_pred - y) ** 2).mean()

    # 3. Backward
    loss.backward()

    # 4. Step (update parameters)
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

    # 5. Zero gradients
    w.grad.zero_()
    b.grad.zero_()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss {loss.item()}")
```

## Same Training Loop Using Optimizer

```python

x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y = torch.tensor([[3.0], [5.0], [7.0], [9.0]])

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

optimizer = torch.optim.SGD([w, b], lr=0.01)

for epoch in range(100):
    # forward
    y_pred = w * x + b

    # loss
    loss = ((y_pred - y) ** 2).mean()

    # backward
    loss.backward()

    # step
    optimizer.step()

    # zero grad
    optimizer.zero_grad()

    if epoch % 10 == 0:
        print(epoch, loss.item())
```
