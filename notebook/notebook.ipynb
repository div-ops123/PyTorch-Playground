1. Tensors represent data in a numerical way

PyTorch tensors are multi-dimensional arrays, similar to NumPy arrays, but with two key advantages: they can run on GPUs for hardware acceleration and are optimized for automatic differentiation (autograd). 


```python
# Scalar
scalar = torch.tensor(7)
scalar
```

```python
# check dimension
scalar.ndim
```

```python
# Get the Python number within a tensor (only works with one-element tensors)
scalar.item()
```

```python
# Vector
vector = torch.tensor([7, 7])
vector
```

```python
# Check the number of dimensions of vector
vector.ndim
```

```python
# Check shape of vector
vector.shape
```
