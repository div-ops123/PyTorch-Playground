# Key Features and Utilities

- **Gradient Accumulation**: By default, PyTorch accumulates (sums) gradients into the .grad attribute rather than overwriting them. This is why you must call optimizer.zero_grad() before each training step.

- **Disabling Tracking**: Use the torch.no_grad() context manager to temporarily stop Autograd from recording operations, which saves memory and speed during model evaluation (inference).

- **Freezing Parameters**: During fine-tuning, you can set requires_grad=False for specific layers to "freeze" them, preventing their weights from being updated.

- **Higher-Order Gradients**: Autograd can compute second-order derivatives (Hessians) by enabling gradient tracking during the backward pass itself.
