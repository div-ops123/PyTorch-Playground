# **AUtograd**

The chain is:

```text
Prediction
→ Compute error
→ Autograd computes gradients
→ Gradient descent updates weights
→ Model improves
```


# Step-by-step gradient

We know:

```
z = y + 4
y = x * 3
```

So:

```
z = (x * 3) + 4
```

Derivative:

```
dz/dx = 3
```

So:

```
x.grad = 3
```

That is what PyTorch computed automatically.

No manual differentiation needed.

This is **Autograd**.

---

# **`requires_grad` (When to Use)**

This tells PyTorch:

**"Track gradients for this tensor"**

Example:

```python
x = torch.tensor(2.0, requires_grad=True)
```

## ML Example

Weights should require gradients

Because:

Weights must update during training.

---

# **`detach()` — When to Use**

detach means:

**Stop tracking gradients from this point**

Example:

```python
x = torch.tensor(2.0, requires_grad=True)

y = x * 3

z = y.detach()

w = z * 4

w.backward()
```

Now:

Gradients **won't flow back to x**

Because we detached.

---

# **`torch.no_grad()` — When to Use**

Used when:

**You don't want gradients temporarily**

Example:

Inference mode

Why?

Because:

During inference:

* no training
* no gradients needed
* save memory
* faster
