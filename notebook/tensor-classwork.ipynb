# Very Important Rule

In PyTorch:

```
dim=0 → down (rows)
dim=1 → across (columns)
```

Think:

```
dim=0  (vertical ↓)

dim=1  (horizontal →)
```

---

# Shape

```python
X.shape
```

Result:

```
(4, 3)
```

Meaning:

* 4 rows (customers)
* 3 columns (categories)

So:

```
dim=0 → 4 (customers)
dim=1 → 3 (categories)
```

---

# When You Do

```python
torch.sum(X, dim=1)
```

You're saying:

**"Sum across columns" (horizontal)**

So:

Customer A:

```
[200, 500, 100]
```

Sum:

```
200 + 500 + 100 = 800
```

Customer B:

```
[150, 200, 80]
```

Sum:

```
430
```

So result:

```
[800, 430, 520, 510]
```

This is **per customer**.

Because **each row = customer**

---

# When You Do

```python
torch.sum(X, dim=0)
```

You're saying:

**"Sum down rows" (vertical)**

Food column:

```
[200,
 150,
 300,
 50]
```

Sum:

```
700
```

Electronics:

```
[500,
 200,
 100,
 400]
```

Sum:

```
1200
```

Clothing:

```
[100,
 80,
 120,
 60]
```

Sum:

```
360
```

Result:

```
[700, 1200, 360]
```

---

# The Mental Model That Never Fails

Think:

```
dim=0 → vertical operation
dim=1 → horizontal operation
```

---

# Visual Memory Trick

Imagine this:

```
dim=1 → ← →  (across rows)

dim=0 → ↑ ↓  (down columns)
```
